```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import seaborn as sns
import joblib
```

```python
# Importation
df = pd.read_csv("global_food_wastage_dataset.csv")
df.head()
```

```python
df.info()
```

```python
 df.describe()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Visualisation des valeurs aberrantes
plt.figure(figsize=(15, 10))
numeric_cols = ["Total Waste (Tons)", "Economic Loss (Million $)", "Avg Waste per Capita (Kg)", 
                "Population (Million)", "Household Waste (%)"]

for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(y=df[col])
    plt.title(f"Boîte à moustaches de {col}")

plt.tight_layout()
plt.show()
```

```python
# Encodage des variables catégorielles
label_encoders = {}
for col in ["Country", "Food Category"]:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

```python
df.head()
```

```python
# Séparation des features et de la cible
X = df.drop(columns=["Economic Loss (Million $)"])  # Features
y = df["Economic Loss (Million $)"]  # Variable cible
```

```python
# Normalisation des variables numériques
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Sauvegarde du scaler
joblib.dump(scaler, "scaler.pkl")
```

```python
# Séparation en train/test (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Vérification des dimensions
print("Train set size:", X_train.shape, y_train.shape)
print("Test set size:", X_test.shape, y_test.shape)
```

### Construction et Comparaison des Modèles

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.preprocessing import PolynomialFeatures
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Modèles de régression
models = {
    "Linear Regression": LinearRegression(),
    "Ridge": Ridge(alpha=1.0),
    "Lasso": Lasso(alpha=1.0),
    "Elastic Net": ElasticNet(alpha=1.0, l1_ratio=0.5, random_state=42),
    "Polynomial Regression": make_pipeline(PolynomialFeatures(degree=2), LinearRegression()),
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42),
    "SVM (Linear)": SVR(kernel='linear')
}

# Entraînement et évaluation des modèles
results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    results[name] = {
        "MSE": mean_squared_error(y_test, y_pred),
        "MAE": mean_absolute_error(y_test, y_pred),
        "R2": r2_score(y_test, y_pred)
    }
```

```python
# Affichage des performances
results_df = pd.DataFrame(results).T
results_df
```

```python
# Visualisation des performances
plt.figure(figsize=(14, 7))

# Graphique R²
plt.subplot(1, 2, 1)
sns.barplot(x=results_df.index, y=results_df["R2"], palette="viridis")
plt.xticks(rotation=45)
plt.title("Comparaison du score R²")
plt.ylabel("R² Score")
plt.ylim(0, 1)

# Graphique MSE
plt.subplot(1, 2, 2)
sns.barplot(x=results_df.index, y=results_df["MSE"], palette="magma")
plt.xticks(rotation=45)
plt.title("Comparaison du MSE (échelle logarithmique)")
plt.yscale('log')
plt.ylabel("MSE (log scale)")

plt.tight_layout()
plt.show()
```

```python
from joblib import dump

# 1. Récupérer le modèle Lasso entraîné
lasso_model = models["Lasso"]

# 2. Sauvegarder le modèle dans un fichier
dump(lasso_model, 'lasso_model.joblib') 

print("Modèle Lasso sauvegardé sous 'lasso_model.joblib'")
```
